```python
read_our_files = lambda data_list : pd.read_csv(data_list, sep = ',', low_memory = False)
# creating a function to read our files so we don't have to do this individually
# separator is a comma because we're reading csvs
# using a lambda function instead of the regular format in order to save space, and because it looks nicer
# low memory set to false so that it takes less time to run
# using read csv
# naming the function read_our_files


# Could also do

# def read_our_files(data_list) :
#     pd.read_csv(data_list, sep = ',', low_memory = False)
```
